**Step 1 — Drop columns by role**
- **PII** (9): Customer Fname, Lname, Email, Password, Street, City, Country, State, Zipcode. No predictive value and must not be shared.
- **DROP** (4): Order Zipcode (86% missing), Product Description (100% missing), Product Image (URL only), Order Profit Per Order (duplicate of Benefit per order).
- **Redundant IDs** (3): Order Customer Id, Order Item Cardprod Id, Product Category Id — keep Customer Id, Product Card Id, Category Id instead.

**Step 2 — Rows**  
Drop any row with missing `Late_delivery_risk` (none expected).

**Step 3 — Dates**  
Parse `order date (DateOrders)` and `shipping date (DateOrders)` to datetime; invalid values become NaT.

**Step 4 — Derived**  
Create `Delay_Gap = Days for shipping (real) - Days for shipment (scheduled)` (positive = late).

```python
# Apply cleaning from the data dictionary: drop PII, DROP, and redundant IDs.
# We keep TARGET, FEATURE, and one of each ID pair; then parse dates and create Delay_Gap.

import pandas as pd

# Columns to remove (PII + DROP + redundant)
COLS_PII = [
    'Customer Fname', 'Customer Lname', 'Customer Email', 'Customer Password',
    'Customer Street', 'Customer City', 'Customer Country', 'Customer State', 'Customer Zipcode'
]
COLS_DROP = ['Order Zipcode', 'Product Description', 'Product Image', 'Order Profit Per Order']
COLS_REDUNDANT_ID = ['Order Customer Id', 'Order Item Cardprod Id', 'Product Category Id']

to_drop = [c for c in COLS_PII + COLS_DROP + COLS_REDUNDANT_ID if c in df_raw.columns]
df = df_raw.drop(columns=to_drop, errors='ignore')

# Duplicates: one row per order line (Order Item Id); drop exact duplicate rows if any
n_before = len(df)
df = df.drop_duplicates()
if len(df) < n_before:
    print(f"Dropped {n_before - len(df)} duplicate rows.")

# Drop rows with missing target (should be none; safety check)
if df['Late_delivery_risk'].isna().any():
    before = len(df)
    df = df.dropna(subset=['Late_delivery_risk'])
    print(f"Dropped {before - len(df)} rows with missing Late_delivery_risk.")

# Parse dates (mixed formats in dataset; coerce errors to NaT)
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'], errors='coerce')
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'], errors='coerce')

# Derived feature: delay gap (real - scheduled). Positive = shipped later than promised.
df['Delay_Gap'] = df['Days for shipping (real)'] - df['Days for shipment (scheduled)']

print(f"Data cleaning complete. Current shape: {df.shape}")
df.info()
```
