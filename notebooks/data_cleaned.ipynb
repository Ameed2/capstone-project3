### 3.5. Check for impossible or suspect values

**Consideration Note:** Why have you chose a profit ratio [extreme loss] (edge case) to be -1, not -0.5? not -1.5?


**Consideration Note:** Why have you chose a profit ratio [impossbile gain] (edge case) to be 1, not 0.5? not 1.5?

```python
print('Impossible value checks:')
print(f'  Days shipping (real) < 0 ..........: {(df["Days for shipping (real)"] < 0).sum():,}')
print(f'  Days scheduled < 0 ................: {(df["Days for shipment (scheduled)"] < 0).sum():,}')
print(f'  Order Item Quantity < 1 ...........: {(df["Order Item Quantity"] < 1).sum():,}')
print(f'  Order Item Discount < 0 ...........: {(df["Order Item Discount"] < 0).sum():,}')
print(f'  Profit Ratio < -1 (extreme loss) ..: {(df["Order Item Profit Ratio"] < -1).sum():,}')
print(f'  Profit Ratio > 1 (impossible gain) : {(df["Order Item Profit Ratio"] > 1).sum():,}')
print(f'  Sales < 0 .........................: {(df["Sales"] < 0).sum():,}')

print()
print('DECISION - Profit Ratio < -1 (6,301 rows, ~3.5% of data):')
print('  These orders recorded a loss larger than the product price itself.')
print('  This is unusual but can reflect heavy discounting, returns, or write-offs.')
print('  SAFE FIX: create a binary indicator flag to preserve the signal rather than')
print('  silently deleting rows. The model can learn from extreme loss events.')
```

### 5.1. Engineer the regression target: `delay_gap`

```python
# delay_gap is the core regression target for Stage 3 of the pipeline.
# Formula: actual days taken minus scheduled days promised.
#   Positive value = arrived late (by that many days)
#   Zero           = exactly on schedule
#   Negative       = arrived early (advance shipping)

df['delay_gap'] = df['shipping_days_actual'] - df['shipping_days_scheduled']

print('delay_gap distribution (days late or early):')
print(df['delay_gap'].value_counts().sort_index().rename('Count').to_frame().to_string())
print(f'\nMean delay  : {df["delay_gap"].mean():.2f} days')
print(f'Pct late (>0): {(df["delay_gap"] > 0).mean()*100:.1f}%')
print(f'Pct early (<0): {(df["delay_gap"] < 0).mean()*100:.1f}%')
```
