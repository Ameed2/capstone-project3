### Dropping useless columns

```python
# --- Drop useless / PII columns ---
df.drop(columns=[
    "Product Description",  # 0 non-null
    "Customer Email", "Customer Password", "Customer Fname", "Customer Lname", "Customer Street",  # Personal Info
    "Product Image",        # URL, not useful
    "Order Zipcode",        # 86% missing
], inplace=True)
```

### Adjust columns types

```python
# --- Convert dates ---
df["order date (DateOrders)"] = pd.to_datetime(df["order date (DateOrders)"])
df["shipping date (DateOrders)"] = pd.to_datetime(df["shipping date (DateOrders)"])
```

```python
# --- Convert binary flags to category ---
df["Late_delivery_risk"] = df["Late_delivery_risk"].astype("category")
df["Product Status"] = df["Product Status"].astype("category")
```

```python
# --- Fix zip codes (float → nullable integer) ---
df["Customer Zipcode"] = df["Customer Zipcode"].astype("Int64").astype("category")  # capital I = nullable
```

```python
# --- Convert low-cardinality text columns to category ---
cat_cols = [
    "Type", "Delivery Status", "Shipping Mode",
    "Customer Segment", "Market", "Order Status",
    "Order Region", "Department Name", "Category Name", "Order State"
]
df[cat_cols] = df[cat_cols].astype("category")
```

```python
# --- Convert ID columns to category ---
id_cols = [
    "Category Id", "Customer Id", "Department Id",
    "Order Customer Id", "Order Id", "Order Item Cardprod Id",
    "Order Item Id", "Product Card Id", "Product Category Id"
]
df[id_cols] = df[id_cols].astype(str)
```

### Checking Redundunt Columns

```python
print((df["Order Customer Id"] == df["Customer Id"]).all())
print((df["Order Item Cardprod Id"] == df["Product Card Id"]).all())
print((df["Product Category Id"] == df["Category Id"]).all())
print((df["Benefit per order"] == df["Order Profit Per Order"]).all())
print((df["Product Price"] == df["Order Item Product Price"]).all())

```

### Drop Redundunt Columns

```python
df.drop(columns=[
    "Order Customer Id",        # same as Customer Id
    "Order Item Cardprod Id",   # same as Product Card Id
    "Product Category Id",      # same as Category Id
    "Order Profit Per Order",   # same as Benefit per order
    "Order Item Product Price", # same as Product Price
], inplace=True)
```

```python
df.rename(columns={
    # Dates
    "order date (DateOrders)"       : "order_date",
    "shipping date (DateOrders)"    : "shipping_date",

    # Delivery & Shipping
    "Days for shipping (real)"      : "shipping_days_actual",
    "Days for shipment (scheduled)" : "shipping_days_scheduled",
    "Late_delivery_risk"            : "late_delivery_risk",
    "Delivery Status"               : "delivery_status",
    "Shipping Mode"                 : "shipping_mode",

    # Order Info
    "Order Id"                      : "order_id",
    "Order Status"                  : "order_status",
    "Order Region"                  : "order_region",
    "Order State"                   : "order_state",
    "Order City"                    : "order_city",
    "Order Country"                 : "order_country",
    "Order Item Id"                 : "order_item_id",
    "Order Item Quantity"           : "item_quantity",
    "Order Item Discount"           : "item_discount_amount",
    "Order Item Discount Rate"      : "item_discount_rate",
    "Order Item Profit Ratio"       : "item_profit_ratio",
    "Order Item Total"              : "item_total",

    # Financials — 
    "Benefit per order"             : "profit_per_order",
    "Sales per customer"            : "customer_total_sales",
    "Sales"                         : "item_sales",
    "Product Price"                 : "product_price",  

    # Customer
    "Customer Id"                   : "customer_id",
    "Customer Segment"              : "customer_segment",
    "Customer City"                 : "customer_city",
    "Customer Country"              : "customer_country",
    "Customer State"                : "customer_state",
    "Customer Zipcode"              : "customer_zipcode",

    # Product — "Product Category Id" already dropped
    "Product Card Id"               : "product_id",
    "Product Name"                  : "product_name",
    "Product Status"                : "product_availability",
    "Category Id"                   : "category_id",
    "Category Name"                 : "category_name",

    # Store/Department
    "Department Id"                 : "department_id",
    "Department Name"               : "department_name",
    "Market"                        : "market",
    "Type"                          : "transaction_type",

    # Location
    "Latitude"                      : "store_latitude",
    "Longitude"                     : "store_longitude",
}, inplace=True)
```

### Missing values analysis (counts and percentages)

Compute the number and percentage of missing values per column, sorted from most to least missing, to guide how we handle them.

```python
# Missing values: counts and percentages
null_counts = df.isna().sum()
null_pct = (null_counts / len(df) * 100).round(2)

missing_summary = (
    pd.DataFrame({
        "null_count": null_counts,
        "null_pct": null_pct,
    })
    .sort_values("null_count", ascending=False)
)

missing_summary.head(20)
```

### Handle missing values
Our dataset is mostly cleaned, we have just 3 missing values in customer_zipcode. Because they are just 3, we can impute it with the most frequent zipcode value. - Maybe this column will not be useful for our analysis

```python
# Getting the most frequent zipcode then fill the null with it
most_frequent_zip = df["customer_zipcode"].mode(dropna=True)[0]
df["customer_zipcode"] = df["customer_zipcode"].fillna(most_frequent_zip)

# Quick re-check of remaining missing values
df.isna().sum().sort_values(ascending=False).head(20)
```

### 9. Outliers and inconsistent values

Detect and treat outliers and inconsistent values (e.g. negative quantities or prices, inconsistent delivery status labels).

```python
# Simple example rules for outliers / inconsistencies

# Remove rows with negative quantities or prices (if such columns exist)
if "order_item_quantity" in df.columns:
    df = df[df["order_item_quantity"] >= 0]

if "order_item_product_price" in df.columns:
    df = df[df["order_item_product_price"] >= 0]

# Standardize delivery status text if present
if "delivery_status" in df.columns:
    df["delivery_status"] = df["delivery_status"].astype(str).str.strip().str.lower()

# You can add more domain-specific rules here
```

### 10. Feature cleaning and engineering

Optionally drop clearly unused columns and create new, more informative features (such as shipping delay and a late-delivery flag).

```python
# Example feature engineering

# Shipping delay in days, if the relevant columns exist
if "days_for_shipping_real" in df.columns and "days_for_shipment_scheduled" in df.columns:
    df["shipping_delay_days"] = (
        df["days_for_shipping_real"] - df["days_for_shipment_scheduled"]
    )

# Binary late-delivery flag based on delay
if "shipping_delay_days" in df.columns:
    df["is_late_delivery"] = (df["shipping_delay_days"] > 0).astype(int)

# Optionally drop clearly unused text / ID columns
cols_to_drop = [
    "product_description",
    "product_image",
    "customer_email",
]

existing_to_drop = [c for c in cols_to_drop if c in df.columns]
print("Dropping columns:", existing_to_drop)

df = df.drop(columns=existing_to_drop)
```
