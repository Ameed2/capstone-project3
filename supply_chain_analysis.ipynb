### Missing values


```python
df.isna().sum().sort_values(ascending=False).head(20)
```

This output lists the columns with the most missing values and how many missing entries they contain. This dataset does not have noticeable number of null values. 
Two columns just have significant number of null values, but it seems that both of them will not benifit us in out model. so dropping them could be a solution.

```python
# Standardize column names to snake_case (lowercase with underscores)

df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(r"[^0-9a-zA-Z]+", "_", regex=True)
      .str.replace(r"_+", "_", regex=True)
      .str.strip("_")
)

df.columns
```

### Univariate analysis – numeric columns

This section looks at each numeric feature one by one using histograms and summary statistics to understand its distribution, central tendency, and spread.

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = df.select_dtypes(include=["number"]).columns

for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col].dropna(), kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

    display(df[col].describe())
    print("\n" + "=" * 80 + "\n")
```

```python
categorical_cols = df.select_dtypes(exclude=["number"]).columns

for col in categorical_cols:
    print(f"Value counts for {col} (top 20):")
    display(df[col].value_counts().head(20))
    print("\n" + "=" * 80 + "\n")
```
